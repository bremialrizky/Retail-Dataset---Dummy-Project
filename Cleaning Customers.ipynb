```python
#impor csv ke pandas
import pandas as pd

cust = pd.read_csv(r'D:\Bahan Belajar\Project\Retail_dataset\customers.csv', sep=';')

cust
```

```python
#Jumlah data kosong
cust.isna().sum()
```

```python
#Cek nilai duplikat di table
cust['customer_id'].value_counts()[cust['customer_id'].value_counts()>1]

cust[cust.duplicated(subset='customer_id', keep = False)].sort_values('customer_id')
```

```python
#Hapus duplikat di kolom seperti order id
cust_copy = cust.copy()
cust_copy = cust_copy.drop_duplicates(subset=['customer_id'])

print('Banyak baris awal:                 :', len(cust))
print('Banyak baris setelah drop duplikat :', len(cust_copy))
print('Banyak baris yang di hapus         :', len(cust) - len(cust_copy))
```

```python
cust = cust_copy
cust
```
