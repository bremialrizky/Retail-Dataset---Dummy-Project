```python
#impor csv ke pandas

import pandas as pd

prod = pd.read_csv(r'D:\Bahan Belajar\Project\Retail_dataset\Product.csv', sep=';')

prod
```

```python
#Jumlah data kosong
prod.isna().sum()
```

```python
prod[prod['product_height_cm'].isna()]
```

```python
#Mengganti isi data
prod['product_category_name'] = prod['product_category_name'].fillna('Uncategorized')
```

```python
#Ganti data yang Nan nilai detail productnya menjadi nilai mean
Nan_Data = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

for Col in Nan_Data:
    prod[Col] = prod.groupby('product_category_name')[Col].transform(lambda x: x.fillna(x.mean()))
```

```python
#Jumlah data di kolom
prod['product_category_name'].value_counts()
```

```python
prod.to_csv("C_Product.csv", index=False)
```
